2.1 Gestion des valeurs manquantes

```python
missing_values = df.isnull().sum()
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].median(), inplace=True)
```

```python
X = df.drop('smoking', axis=1)
y=df ['smoking']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Nous avons utilisé 80% des données pour l'entraînement et 20% pour le test.

Il faut diviser les données afin de créer un ensemble de tests indépendant pour évaluer les performances du modèle. Cela permet de vérifier si le modèle généralise bien des données qu'il n'a pas vues lors de l'entraînement.

```python
from sklearn import tree
from sklearn.model_selection import GridSearchCV
error_score='raise'
clf = tree.DecisionTreeClassifier(random_state=0)
X_train_1 , X_test_1 ,y_train_1,y_test_1=train_test_split(X,y,test_size=0.2,random_state=100)
distributions = {'criterion': ["gini","entropy","log_loss"],
    'splitter': ['best', 'random']
    }

CLF = GridSearchCV(clf, distributions)
search = CLF.fit(X_train,y_train)
search.best_params_
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
clf = tree.DecisionTreeClassifier(random_state=79,splitter='best',criterion = 'entropy')
CLF.fit(X_train_1,y_train_1)
y_pred_1 = CLF.predict(X_test_1)
confusion_mat_result_1 = confusion_matrix(y_pred_1,y_test_1)
confusion_mat_result_1
sns.heatmap(confusion_mat_result_1, annot=True)
```
